```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Function to load images from a directory
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        img = cv2.imread(img_path)
        if img is not None:
            images.append(img)
    return images
```

```python
def resize_images(images, target_size):
    resized_images = []
    for img in images:
        resized_img = cv2.resize(img, target_size)
        resized_images.append(resized_img)
    return np.array(resized_images)

# Define the target size for resizing
target_size = (350, 350)  # Define the common height and width

# Resize train_images and valid_images to a common size
train_images_resized = resize_images(train_images, target_size)
valid_images_resized = resize_images(valid_images, target_size)
```

```python
# Convert lists of images to numpy arrays
train_images = np.array(train_images_resized)
valid_images = np.array(valid_images_resized)
```

```python
def normalize_images_in_batches(images_folder, batch_size=100):
    num_images = len(os.listdir(images_folder))
    for i in range(0, num_images, batch_size):
        batch_images = load_and_resize_images(images_folder, start_index=i, batch_size=batch_size)
        normalized_batch = normalize_images(batch_images)
        save_normalized_images(normalized_batch, folder='normalized_images', start_index=i)

# Function to load and resize a batch of images from a directory
def load_and_resize_images(folder, start_index, batch_size, target_size=(350, 350)):
    images = []
    filenames = sorted(os.listdir(folder))
    for j in range(start_index, min(start_index + batch_size, len(filenames))):
        img_path = os.path.join(folder, filenames[j])
        img = cv2.imread(img_path)
        if img is not None:
            img = cv2.resize(img, target_size)
            images.append(img)
    return images

# Function to save a batch of normalized images
def save_normalized_images(images, folder, start_index):
    for j, img in enumerate(images):
        cv2.imwrite(f'{folder}/image_{start_index + j}.png', img)

# Function to normalize a batch of images
def normalize_images(images):
    normalized_images = np.array(images).astype('float32') / 255.0  # Convert list to numpy array and normalize
    return normalized_images

# Normalize train_images, valid_images, and watermark_image using batch size normalization
normalize_images_in_batches(train_folder)
normalize_images_in_batches(valid_folder)
```

```python
def change_color(image, hue_shift=0, saturation_scale=1.0, value_scale=1.0):
    """
    Change the color of the image by adjusting the hue, saturation, and value channels in the HSV color space.
    
    Args:
        image: Input image (grayscale or RGB format).
        hue_shift: Shift in hue value (range: -180 to 180).
        saturation_scale: Scaling factor for saturation (range: 0.0 to infinity).
        value_scale: Scaling factor for value (brightness) (range: 0.0 to infinity).
    
    Returns:
        Transformed image.
    """
    if len(image.shape) == 2:  # Grayscale image
        # Convert grayscale image to RGB
        image_rgb = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    elif len(image.shape) == 3 and image.shape[2] == 3:  # RGB image
        image_rgb = image
    else:
        raise ValueError("Unsupported image format. Expected grayscale or RGB image.")
    
    # Convert RGB image to HSV color space
    hsv_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)
    
    # Shift hue channel
    hsv_image[:, :, 0] = (hsv_image[:, :, 0] + hue_shift) % 180
    
    # Scale saturation channel
    hsv_image[:, :, 1] = np.clip(hsv_image[:, :, 1] * saturation_scale, 0, 255).astype(np.uint8)
    
    # Scale value channel
    hsv_image[:, :, 2] = np.clip(hsv_image[:, :, 2] * value_scale, 0, 255).astype(np.uint8)
    
    # Convert back to RGB color space
    return cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB)

# Example usage:
# Apply color change transformation to train_images and valid_images
hue_shift = 20
saturation_scale = 1.5
value_scale = 1.0

transformed_train_images = [change_color(image, hue_shift, saturation_scale, value_scale) for image in train_images]
transformed_valid_images = [change_color(image, hue_shift, saturation_scale, value_scale) for image in valid_images]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Define the model architecture for watermark extraction
extraction_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(350, 350, 3)),  # Input shape is (height, width, channels)
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')  # Output layer with sigmoid activation for binary classification
])

# Compile the model
extraction_model.compile(optimizer='adam',
                         loss='binary_crossentropy',
                         metrics=['accuracy'])

# Train the model
extraction_history = extraction_model.fit(transformed_train_images, train_labels, epochs=10, batch_size=32,
                                          validation_data=(transformed_valid_images, val_labels))

# Evaluate model on validation data
extraction_loss, extraction_accuracy = extraction_model.evaluate(transformed_valid_images, val_labels)
print("Validation Loss:", extraction_loss)
print("Validation Accuracy:", extraction_accuracy)
```
